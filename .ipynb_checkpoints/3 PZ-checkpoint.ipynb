```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y = [i**2 for i in x]
plt.figure(figsize=(5, 5))
plt.title("Квадратичная зависимость y = x^2")
plt.xlabel("x") 
plt.ylabel("y")
plt.grid()
plt.plot(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=1, label=f'Среднее: {data.mean():.2f}')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Гистограмма нормального распределения')
plt.legend()
plt.grid(True)
plt.show()
```

***Задание №7***

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
ax.set_title('3D-график функции $z = \sin(\sqrt{x^2 + y^2})$')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создание сетки значений
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Построение 3D-графика
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Исправленная строка с raw-строкой (r перед строкой)
ax.set_title(r'3D-график функции $z = \sin(\sqrt{x^2 + y^2})$', fontsize=14)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

# Добавление цветовой шкалы
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

***Задание №9***

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)  
matrix = np.random.rand(10, 10)
plt.figure(figsize=(8, 6))
heatmap = plt.imshow(matrix, cmap='viridis')
plt.colorbar(heatmap, label='Значение')
plt.title('Тепловая карта случайной матрицы 10×10')
plt.xticks(range(10))
plt.yticks(range(10))
for i in range(10):
    for j in range(10):
        plt.text(j, i, f'{matrix[i, j]:.2f}', 
                ha='center', va='center', 
                color='white' if matrix[i, j] > 0.5 else 'black')

plt.show()
```

```python
import matplotlib.pyplot as plt

grades = ['2', '3', '4', '5']
students = [5, 12, 20, 13]

plt.figure(figsize=(10, 6))
bars = plt.barh(grades, students, color=['red', 'orange', 'blue', 'green'])

plt.title('Распределение оценок студентов', fontsize=14)
plt.xlabel('Количество студентов', fontsize=12)
plt.ylabel('Оценки', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2, f'{int(width)}', 
             va='center', fontsize=10)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

a, b = 0, np.pi
n = 1000  # Увеличиваем количество точек для точности
x = np.linspace(a, b, n)
y = f(x)
area = np.sum(y) * (b - a)/n

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.fill_between(x, y, color='skyblue', alpha=0.4)
plt.title(f'Площадь под sin(x) ≈ {area:.4f} (численный метод)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Создаем сетку значений
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 - y**2  # Вычисляем значения функции

# Создаем 3D-график
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Рисуем поверхность с цветовой картой
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Настраиваем оформление графика
ax.set_title('Поверхность функции $f(x, y) = x^2 - y^2$', fontsize=14)
ax.set_xlabel('Ось X', fontsize=12)
ax.set_ylabel('Ось Y', fontsize=12)
ax.set_zlabel('Ось Z', fontsize=12)

# Добавляем цветовую шкалу
fig.colorbar(surf, shrink=0.5, aspect=5, label='Значение функции')

# Устанавливаем угол обзора
ax.view_init(elev=30, azim=45)

plt.tight_layout()
plt.show()
```
