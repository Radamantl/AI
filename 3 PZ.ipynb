```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y = [i**2 for i in x]
plt.figure(figsize=(5, 5))
plt.title("Квадратичная зависимость y = x^2")
plt.xlabel("x") 
plt.ylabel("y")
plt.grid()
plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 100)
y1 = x
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]
plt.figure(figsize=(5, 5))
plt.title("Три грфика в одном")
plt.xlabel("x") 
plt.ylabel("y")
plt.grid()
plt.plot(x,y1,x,y2,x,y3)
```

```python
import matplotlib.colors as mcolors
bc = mcolors.BASE_COLORS
x = np.arange(0, 10, 0.20)
y = np.cos(x)
num_set = np.random.randint(1, len(mcolors.BASE_COLORS), len(x))
sizes = num_set * 35
colors = [list(bc.keys())[i] for i in num_set]
plt.scatter(x, y, s=sizes, alpha=0.4, c=colors, linewidths=2,edgecolors="face")
plt.plot(x, y, "g--", alpha=0.4)
```
